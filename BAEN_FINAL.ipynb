```python
# Author: Partho Ghose
import numpy as np
from scipy import stats

# Given data
n = 20                                 # Sample size
x_bar = 725.5                          # Sample mean (mm)
s = 200                                # Sample standard deviation (mm)
mu_0 = 750                             # Claimed mean
alpha = 0.05                           # Significance level for mean test
alpha_ci = 0.10                        # For 90% CI on variance

df = n - 1                             # Degrees of freedom

# ---------------------------------------------
# 📘 Step 1: Hypothesis Test for Population Mean
# ---------------------------------------------
print("📘 Step 1: Hypothesis Test for the Mean")
print("Null Hypothesis: H₀: μ = 750 mm")
print("Alternative Hypothesis: H₁: μ ≠ 750 mm (two-tailed test)")

# t-statistic calculation
t_stat = (x_bar - mu_0) / (s / np.sqrt(n))
t_critical = stats.t.ppf(1 - alpha/2, df)
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

print(f"\nTest statistic (t):")
print(f"t = ({x_bar} - {mu_0}) / ({s} / √{n}) = {t_stat:.4f}")
print(f"Degrees of freedom: {df}")
print(f"Critical t-value at α={alpha}: ±{t_critical:.4f}")
print(f"p-value = {p_value:.4f}")

# Hypothesis test decision
if abs(t_stat) > t_critical:
    print("❌ Reject H₀: The mean rainfall significantly differs from 750 mm.")
else:
    print("✅ Fail to Reject H₀: No significant difference from 750 mm.")

# ----------------------------------------------------
# 📘 Step 2: 90% Confidence Interval for Sample Variance
# ----------------------------------------------------
print("\n📘 Step 2: 90% Confidence Interval for Population Variance")

s_squared = s ** 2  # Sample variance

# Chi-square bounds
chi2_lower = stats.chi2.ppf(alpha_ci / 2, df)
chi2_upper = stats.chi2.ppf(1 - alpha_ci / 2, df)

# Confidence interval for variance
var_lower = (df * s_squared) / chi2_upper
var_upper = (df * s_squared) / chi2_lower

print(f"\ns² = {s}² = {s_squared:.2f}")
print(f"χ² lower (α/2 = {alpha_ci/2:.2f}, df = {df}) = {chi2_lower:.4f}")
print(f"χ² upper (1 - α/2 = {1 - alpha_ci/2:.2f}, df = {df}) = {chi2_upper:.4f}")
print(f"90% Confidence Interval for Variance = [{var_lower:.2f}, {var_upper:.2f}]")
```



## 📘 Why t-test and chi-square test are used

---

###  Part 1: Why **t-test** is used for testing the **mean**

We are testing whether the **mean annual rainfall** in a semi-arid region is 750 mm, based on sample data.

#### ✅ Given:

* Sample mean $\bar{X} = 725.5$ mm
* Sample standard deviation $s = 200$ mm
* Sample size $n = 20$
* Claimed population mean $\mu_0 = 750$ mm
* Population standard deviation $\sigma$ is **unknown**

#### ✅ Reason for using the **t-test**:

* The population standard deviation $\sigma$ is unknown
* The sample size is small ($n < 30$)
* Therefore, we use the **Student’s t-distribution** instead of the normal distribution

####  **Test Statistic Formula**:

$$
t = \frac{\bar{X} - \mu_0}{s / \sqrt{n}}
$$

Where:

* $\bar{X}$: Sample mean
* $\mu_0$: Claimed population mean
* $s$: Sample standard deviation
* $n$: Sample size

We compare this $t$-value with the **critical value** from the t-distribution table (based on $n-1$ degrees of freedom), or compute the **p-value** to test the hypothesis.

---

### 🔹 Part 2: Why **chi-square test** is used for confidence interval of **variance**

We are constructing a **90% confidence interval** for the **population variance** $\sigma^2$, based on sample data.

#### ✅ Given:

* Sample variance $s^2 = 200^2 = 40000$
* Sample size $n = 20$, so degrees of freedom = $n - 1 = 19$

#### ✅ Reason for using the **chi-square distribution**:

* When the population is assumed to be **normally distributed**, the sample variance follows a **chi-square distribution**:

$$
\chi^2 = \frac{(n - 1)s^2}{\sigma^2}
$$

This formula allows us to rearrange and construct a confidence interval for the unknown population variance $\sigma^2$.

####  **Confidence Interval Formula**:

$$
\left[ \frac{(n-1)s^2}{\chi^2_{upper}}, \frac{(n-1)s^2}{\chi^2_{lower}} \right]
$$

Where:

* $\chi^2_{lower}$: Chi-square critical value at $\alpha/2$
* $\chi^2_{upper}$: Chi-square critical value at $1 - \alpha/2$
* $s^2$: Sample variance
* $n-1$: Degrees of freedom

---

### ✅ Summary Table

| Purpose                          | Distribution Used  | Why it is used                                    |
| -------------------------------- | ------------------ | ------------------------------------------------- |
| Test mean with unknown σ         | **t-distribution** | Sample size is small, population σ unknown        |
| Confidence interval for variance | **chi-square**     | Variance follows χ² under assumption of normality |

### 📘 What is Multiple Linear Regression (MLR)?

**Multiple Linear Regression (MLR)** is a statistical technique used to model the relationship between **one dependent variable** (output) and **two or more independent variables** (inputs or predictors).

It fits a linear equation of the form:

$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \cdots + \beta_n X_n + \varepsilon
$$

**Where:**

* $Y$: Dependent variable (e.g., direct runoff)
* $X_1, X_2, \dots, X_n$: Independent variables (e.g., stream length, drainage density)
* $\beta_0$: Intercept (constant term)
* $\beta_1, \beta_2, \dots$: Coefficients (slopes)
* $\varepsilon$: Error term

The goal is to find the **best-fitting line (or plane/hyperplane)** that minimizes the difference (error) between the predicted and actual observed values of $Y$.



```python
import numpy as np
import pandas as pd

# Step 1: Define the dataset with actual variable names
data = {
    'Length of Longest Stream (km)': [20.02, 15.22, 17.43, 8.18, 10.17, 9.82, 19.03, 14.60, 19.84, 18.89],
    'Drainage Density (km/km²)': [14.42, 14.95, 14.87, 16.31, 16.00, 16.29, 16.50, 16.81, 13.97, 13.86],
    'Direct Runoff (cm)': [10.8, 8.5, 8.4, 6.8, 6.1, 7.0, 8.6, 8.0, 8.3, 9.8]
}

df = pd.DataFrame(data)
n = len(df)

# Step 2: Create design matrix X and output vector Y
X0 = np.ones(n)
X1 = df['Length of Longest Stream (km)'].values
X2 = df['Drainage Density (km/km²)'].values
Y = df['Direct Runoff (cm)'].values.reshape(-1, 1)
X = np.column_stack((X0, X1, X2))  # Final design matrix

# print("\n✅ Step 2: Design Matrix (X) and Output Vector (Y)")
# print(pd.DataFrame(X, columns=["Intercept", "Length of Longest Stream (km)", "Drainage Density (km/km²)"]))
# print("\nY = Direct Runoff (cm):\n", pd.DataFrame(Y, columns=["Direct Runoff (cm)"]))

# Step 3: Compute coefficients β = (XᵀX)^-1 XᵀY
Xt = X.T
XtX = Xt @ X
XtY = Xt @ Y
XtX_inv = np.linalg.inv(XtX)
beta = XtX_inv @ XtY
b0, b1, b2 = beta.flatten()

print("\n✅ Step 3: Regression Coefficients")
print("β = (XᵀX)^-1 XᵀY")
print(f"β₀ (Intercept) = {b0:.4f}")
print(f"β₁ (Length of Longest Stream) = {b1:.4f}")
print(f"β₂ (Drainage Density) = {b2:.4f}")

# Dynamically format the regression equation
def format_term(coef, name):
    return f" {'+' if coef >= 0 else '-'} {abs(coef):.4f} * {name}"

equation = f"Runoff (cm) = {b0:.4f}" + \
           format_term(b1, 'Length of Longest Stream (km)') + \
           format_term(b2, 'Drainage Density (km/km²)')

print("\n📘 Regression Equation:")
print(equation)

# Step 4: Predict runoff
Y_hat = X @ beta
df['Predicted Runoff (cm)'] = Y_hat
print("\n✅ Step 4: Predicted Runoff for Each Basin")
print(df[['Length of Longest Stream (km)', 'Drainage Density (km/km²)', 'Direct Runoff (cm)', 'Predicted Runoff (cm)']])

# Step 5: Residuals and SSE
residuals = Y - Y_hat
SSE = np.sum(residuals ** 2)
df['Residual (cm)'] = residuals

print("\n✅ Step 5: Residuals")
print(df[['Direct Runoff (cm)', 'Predicted Runoff (cm)', 'Residual (cm)']])
print(f"\nSSE (Sum of Squared Errors) = {SSE:.4f}")

# Step 6: Coefficient of Determination (R²)
Y_mean = np.mean(Y)
SST = np.sum((Y - Y_mean) ** 2)
R_squared = 1 - SSE / SST

print("\n✅ Step 6: Coefficient of Determination (R²)")
print(f"Ȳ (Mean of Y) = {Y_mean:.4f}")
print(f"SST (Total Sum of Squares) = {SST:.4f}")
print(f"R² = {R_squared:.4f}")
```

| Code Line                      | Mathematical Form                               | Explanation                                                                                    |
| ------------------------------ | ----------------------------------------------- | ---------------------------------------------------------------------------------------------- |
| `Xt = X.T`                     | $X^\top$                                        | Transpose the design matrix. Converts shape from (n, p) to (p, n) to prepare for dot products. |
| `XtX = Xt @ X`                 | $X^\top X$                                      | Computes the **Gram matrix**, which summarizes the variance and correlation of the predictors. |
| `XtY = Xt @ Y`                 | $X^\top Y$                                      | Projects the output vector onto each predictor — how much each variable "aligns" with Y.       |
| `XtX_inv = np.linalg.inv(XtX)` | $(X^\top X)^{-1}$                               | Inverts the Gram matrix — required for solving normal equations.                               |
| `beta = XtX_inv @ XtY`         | $\boldsymbol{\beta} = (X^\top X)^{-1} X^\top Y$ | Solves the **normal equation**, giving the best-fit coefficients that minimize squared error.  |

```python
# River Stage Normality Analysis using Skewness Test
# This script performs step-by-step calculations to convert streamflow to river stage,
# and test normality using skewness at 5% significance level.

import numpy as np
from scipy.stats import skew

def skewness_normality_analysis(streamflow_1000):

    # print("Step 1: Convert Streamflow to m³/s")
    # print("Q_i (m³/s) = Q_i (1000 m³/s) × 1000")
    # Q = np.array(streamflow_1000) * 1000
    # print("Q =", Q.tolist(), "\n")

    # print("Step 2: Calculate River Stage (G) using the rating curve:")
    # print("G_i = sqrt(Q_i / 59.5) + 5")
    # G = np.sqrt(Q / 59.5) + 5
    # print("G =", G.tolist(), "\n")

    print("Step 1: Convert Streamflow to m³/s")
    print("Q_i (m³/s) = Q_i (1000 m³/s) × 1000")
    Q = np.array(streamflow_1000) * 1000
    print("Q =", Q[:10].tolist())
    print( Q[10:20].tolist())
    print( Q[20:].tolist(), "\n")

    print("Step 2: Calculate River Stage (G) using the rating curve:")
    print("G_i = sqrt(Q_i / 59.5) + 5")
    G = np.sqrt(Q / 59.5) + 5
    print( "G =", G[:10].round(3).tolist())
    print( G[10:20].round(3).tolist())
    print( G[20:].round(3).tolist(), "\n")



    print("Step 3: Calculate Mean (Ḡ):")
    G_mean = np.mean(G)
    print("Ḡ = (1/n) × ΣG_i =", G_mean, "\n")

    print("Step 4: Calculate Sample Standard Deviation (s):")
    G_std = np.std(G, ddof=1)
    print("s = sqrt((1/(n-1)) × Σ(G_i - Ḡ)²) =", G_std, "\n")

    print("Step 5: Calculate Skewness using:")
    print("Skewness = (1/n) × Σ[((G_i - Ḡ)/s)^3]")
    skewness_val = skew(G, bias=False)
    print("Skewness =", skewness_val, "\n")

    print("Step 6: Calculate Standard Error of Skewness (SE):")
    n = len(G)
    SE = np.sqrt((6 * n * (n - 1)) / ((n - 2) * (n + 1) * (n + 3)))
    print("SE = sqrt((6n(n-1))/((n-2)(n+1)(n+3))) =", SE, "\n")

    print("Step 7: Calculate Test score for Skewness:")
    z = skewness_val / SE
    print("S = Skewness / SE =", z, "\n")

    print("Step 8: Conclusion at 5% significance level (α = 0.05):")
    print(" At 5% significance level (α = 0.05):  z_(α/2)*sqrt(6/n) value =0.8765 (using online calculator and equation 9.32 from book) ")
    if abs(z) <= 1.96:
        print("Since |S| =", abs(z), "≤ 1.047 → Fail to reject H₀ → Data is normally distributed.\n")
    else:
        print("Since |S| =", abs(z), "> 1.047 → Reject H₀ → Data is NOT normally distributed.\n")

    return {
        "Streamflow (m³/s)": Q.tolist(),
        "River Stage (m)": G.tolist(),
        "Mean": G_mean,
        "Std Dev": G_std,
        "Skewness": skewness_val,
        "SE": SE,
        "z-score": z,
        "Normal (5%)": abs(z) <= 1.047
    }

# Example streamflow data in 1000 m³/s
streamflow_data_1000 = [
    14.12, 22.05, 22.34, 20.07, 21.15, 19.82, 20.65, 23.57, 22.19, 18.32,
    14.45, 12.72, 13.67, 12.58, 9.33, 9.67, 10.65, 14.47, 11.02, 9.82,
    12.50, 16.10, 17.40, 9.48, 8.41, 9.33, 10.40, 12.62, 15.30, 13.84
]

# Run the analysis
results = skewness_normality_analysis(streamflow_data_1000)
```
