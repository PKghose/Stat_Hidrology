```python
import numpy as np
import pandas as pd

# Step 1: Create DataFrame
data = {
    "Precipitation": [0.00, 0.03, 5.99, 4.97, 9.50, 3.94, 145.06, 241.36, 413.60, 216.46, 41.37, 44.43, 0.81, 1.94, 6.50, 1.56, 0.57, 2.50, 67.74, 422.32, 370.69, 237.83, 210.00, 0.00],
    "Precipitable_Water": [11.15, 12.39, 15.54, 15.59, 20.99, 21.41, 36.46, 49.36, 49.66, 41.48, 31.27, 20.99, 20.30, 18.51, 21.08, 17.42, 23.29, 23.74, 43.73, 53.09, 51.05, 48.86, 39.29, 18.09]
}
df = pd.DataFrame(data)

# Step 2: Create X and Y
Y = df["Precipitation"].values.reshape(-1, 1)
X = df["Precipitable_Water"].values.reshape(-1, 1)
X = np.hstack((np.ones_like(X), X))  # Add intercept

# Step 3: Compute X^T X
XtX = X.T @ X
print("X^T X:\n", XtX)

# Step 4: Compute X^T Y
XtY = X.T @ Y
print("\nX^T Y:\n", XtY)

# Step 5: Compute (X^T X)^-1
XtX_inv = np.linalg.inv(XtX)
print("\n(X^T X)^-1:\n", XtX_inv)

# Step 6: Compute beta coefficients
beta = XtX_inv @ XtY
intercept, slope = beta.flatten()
print("\nEstimated Coefficients (β):\n", beta)
print(f"\n✅ Final Model: Ŷ = {intercept:.4f} + {slope:.4f} * X")

# Step 7: Predict Y
Y_pred = X @ beta
print("\nPredicted Y (Ŷ):\n", Y_pred.flatten())

# Step 8: Calculate Residuals
residuals = Y - Y_pred
print("\nResiduals (Y - Ŷ):\n", residuals.flatten())

# Step 9: SSE (Sum of Squared Errors)
sse = np.sum(residuals ** 2)
print("\nSSE (Sum of Squared Errors):", sse)

# Step 10: R² Calculation
ss_total = np.sum((Y - np.mean(Y)) ** 2)
r2 = 1 - sse / ss_total
print("R² (Coefficient of Determination):", r2)
```
